#1. Attributes (e.g., df.shape)
Definition: An attribute is a property or a characteristic of an object. It represents a piece of data that is associated with the object. Attributes do not perform any action; they simply store information.
Does not require parentheses: Since attributes just hold data, you don’t need parentheses to access them.
Example:
df.shape: This attribute gives you the dimensions of the DataFrame, returning a tuple (rows, columns). It's a static piece of information stored in the DataFrame object.

2. Methods (e.g., df.describe())
Definition: A method is a function that is associated with an object and can perform actions on that object. Methods often take arguments, perform calculations or operations, and can return a result.
Requires parentheses: Since methods are functions, you need to call them with parentheses, even if there are no arguments. The parentheses signify the execution of the method.
Example:
df.describe(): This method computes summary statistics for the DataFrame’s columns. When you call it, the method performs the computation and returns the result.

Key Differences:
Attributes are like static properties or data points (e.g., df.shape, df.columns, df.index), while
Methods are functions that perform operations on the object (e.g., df.describe(), df.mean(), df.drop()).

In summary:
Attributes provide stored data and don't need parentheses.
Methods perform actions or calculations and need parentheses to be called.

#1. Count
Definition: The count is the number of non-missing (non-null) values in each column. If a column contains missing values (NaN), those are excluded from the count.
2. Mean
Definition: The mean is the arithmetic average of the values in the column. It is calculated by summing all the values and dividing by the count of non-missing values: Mean=count of values/∑value
3. Standard Deviation (std)
Definition: The standard deviation measures the amount of variation or dispersion of values from the mean. A higher standard deviation indicates that the values are more spread out from the mean, while a lower standard deviation means they are closer to the mean. 
4. Minimum (min)
Definition: The minimum is the smallest value in the column.
5. 25% (First Quartile or Q1)
Definition: This is the value below which 25% of the data falls. It is the first quartile (Q1) of the data and provides insight into the lower end of the distribution.
6. 50% (Median or Second Quartile or Q2)
Definition: The median is the middle value of the data, such that 50% of the values lie below it and 50% above. It is also called the second quartile (Q2). If the data is sorted, it represents the middle value when there is an odd number of observations, or the average of the two middle values when there is an even number.
7. 75% (Third Quartile or Q3)
Definition: This is the value below which 75% of the data falls. It is the third quartile (Q3) of the data and provides insight into the upper end of the distribution.
8. Maximum (max)
Definition: The maximum is the largest value in the column.
Summary:
Count: Number of non-null values.

Min: Smallest value.
25%: First quartile (Q1), value at 25% of data.
50%: Median (Q2), value at the midpoint of data.

#1. Avoid Unnecessary Row Deletion
When a column has many missing values, dropping rows with df.dropna() before deleting the column could result in an unnecessary loss of data from other columns. If you apply df.dropna() first, it will remove rows with missing values across all columns, including the column you might not even need. This can significantly reduce the number of rows, potentially discarding valuable data that could have been kept.

By deleting the problematic column first, you ensure that its missing values don’t affect your decision to drop rows, thus preserving as many rows as possible.

2. Efficiency in Data Cleaning
Removing irrelevant or mostly missing columns using del df['col'] reduces the dataset size before applying df.dropna(). This makes the operation more efficient by working with fewer columns. You can then clean the dataset by dropping rows with missing values in the important columns without being impacted by columns that have already been removed.

3. More Targeted Data Cleaning
By deleting unnecessary columns first, you focus the cleaning process on the columns that matter for your analysis. When you apply df.dropna() afterward, you are only removing rows based on missing values in the columns that are still relevant. This helps to avoid over-deletion of rows due to irrelevant or unnecessary columns.

Example Scenario:
You have a dataset with columns like customer_id, age, gender, purchase_history, and social_media_handle. If social_media_handle has many missing values but is not important for your analysis, applying df.dropna() first could cause you to lose many rows where other important columns (like customer_id or age) have complete data.
